```python
# Matplotlib >>>> it is the library of python , to used for data visulisation .
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = np.array([1 , 3])
ypoint = np.array([2 , 4])
plt.plot(xpoint , ypoint)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

cgpa = np.array([6, 6.5, 7, 7.5, 8, 8.5, 9, 9.9])
package = np.array([5, 5.6, 6, 6.8, 7, 7.8, 8, 8.4])

# plot with style
plt.plot(cgpa, package, '.-b', ms=10, mec="red", linewidth=2.3)

# font settings
font1 = {'family': 'serif', 'color': 'red', 'size': 16}
font2 = {'family': 'serif', 'color': 'black', 'size': 12}

# labels and title with fonts
plt.xlabel("CGPA", fontdict=font2)
plt.ylabel("PACKAGE", fontdict=font2)
plt.title("Shoolini University", fontdict=font1)

plt.grid(True)
plt.show()
```

```python
# Bar plot >>> it is work on numerical as well as categorical data. ex. fruit == apple banana orange >> categorical data . quantity of fruits >>> numerical data.
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple" , "Banana" , "orange" , "Grapes"])
quantity = np.array([30 , 20 , 40 , 10])
color = np.array(["red"  , "yellow" , "orange" , "green"])
plt.bar(fruit , quantity , color=color)
plt.xlabel("Fruits")
plt.ylabel("Quantiy")
plt.title("FRUIT QUANTITY")
plt.show()



```

```python
# bar graph in horizontal direction ....
fruit = np.array(["Apple" , "Banana" , "orange" , "Grapes"])
quantity = np.array([30 , 20 , 40 , 10])
color = np.array(["red"  , "yellow" , "orange" , "green"])
plt.barh(fruit , quantity , color = color)
plt.xlabel("Quantity")
plt.ylabel("Fruits")
plt.title("FRUIT QUANTITY")
plt.show()
```

```python
# Pie Plots >>>>
quantiy = np.array([20 , 30 ,36 , 50 , 40])
fruit = np.array(["Apple" , "Banana" , "orange" , "Grapes" , "Mango"])
myexplode = [0.3 , 0 , 0 , 0 , 0]
plt.pie(quantiy , labels = fruit , explode = myexplode)
plt.legend()
plt.legend(title = "Frits" , loc = "upper left" ) # set legend accordindly with the syntax >>> upper right , upper left , lower right , lower left , best.
plt.show()
```

```python
# scatter plot .
age_of_car = np.array([1 , 4 , 6 , 10 , 20 , 25 ])
speed_of_car = np.array([200 , 100 , 90 , 50 , 40 , 20])
plt.scatter(age_of_car , speed_of_car)
plt.xlabel("AGE")
plt.ylabel("SPEED")
plt.title("AGE and SPEED of Car")
plt.grid(True)
plt.show()
```
